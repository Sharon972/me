## ¿Cómo filtro filas especifícas de un `DataFrame`?

03_subset_rows.svg

**Estoy interesado en pasajeros mayores a 35 años**

```python
above_35 = titanic[titanic["Age"] > 35] #a la columna de nombre Age le digo que son mayores de 35
#titanic["Age"] > 35 es F o V
#al escribir titanic[titanic["Age"] > 35] obtiene el subconjunto de la base de datos titanic que cumple la condición

above_35.head()
```

Para seleccionar filas basadas en una expresión condicional, usamos la condición de selección dentro de los corchetes.

La condición dentro de los corchetes de selección `titanic["Age"] > 35` verifica para que filas la columna `Age` posee valores mayores a 35.

```python
titanic["Age"] > 35
```

La salida de la expresión condicional (`>`, pero también `==`,`!=`, `<`, `<=`, ... funciona) es en realidad una `Series` de pandas de valores de booleanos (o `True` o `False`) con la misma cantidad de filas que el `DataFrame` original. Estas `Series` de valores booleanos pueden ser utilizados para filtrar el `DataFrame` poniendolo en los corchetes de selección. Solo las filas que contengan el valor `True` serán seleccionadas.

Sabemos desde antes que el `DataFrame` original de Titanic consiste de 891 filas. Ahora demos un vistazo a la cantidad de filas que satisfacen la condición utilizando el atributo `shape`

```python
above_35.shape
```

**Estoy interesado en los pasajeros del titanic de las clases de cabinas 2 y 3**

```python
class_23 = titanic[titanic["Pclass"].isin([2, 3])]
# Seleccionamos columana titanic["Pclass"] == 2 , isin( si 2 y 3 esta dentro me da True)
class_23.head()
```

Similarmente a las expresiones de condiciones, la función condicional `isin()` retorna `True` para cada fila cuyo valor este en la lista proveida. Para filtrar las filas basada en esta función, usamos la función condicional dentro de los corchetes de selección `[]`. En este caso, la condición dentro de los corchetes `titanic["Pclass"].isin([2, 3])` verifica en que filas el valor de la columna `Pclass` es 2 o 3

La expresión de arriba es equivalente a filtrar las filas por su clase siendo 2 o 3 y combinando ambas expresiones con el operador or `|`:

```python
class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]
#que sea igual a 2 o que sea igual a 3 las clases de las cabinas, & para "y"

class_23.head()
```

**Cuando combinamos varias expresiones condicionales, cada condición debe estar rodeada por paréntesis `()`. Además, no se pueden utilizar `or`/`and`, sino debemos usar sus operadores `|` y `&` respectivamente.**

Quiero trabajar solo con los datos de los pasajeros cuya edad es conocida

```python
age_no_na = titanic[titanic["Age"].notna()] #notna (funcion condicional me retorna los q no sean nulos)
#no nulos para la columna Age
age_no_na.head()
```

La función condicional `notna()` retorna `True` para cada fila cuyo valores no son nulos. Esto se puede utilizar en conjunto con los corchetes de selección para filtrar los datos de la tabla.

Verificamos entonces cual es nuestra nueva cantidad de filas

```python
age_no_na.shape
```

## ¿Cómo selecciono filas y columnas específicas de un `DataFrame`?

03_subset_columns_rows.svg

**Estoy interesado en los nombres de los pasajeros mayores a 35 años**

```python
adult_names = titanic.loc[titanic["Age"] > 35, "Name"]
#Para seleccionar columnas y filas PARA ELLO USO variable.loc "nombres con edad mayores de 35"
#CONDICIONES TEXTO
adult_names.head()
```

En este caso, el subconjunto de ambas, columnas y filas, en una sola sentada y solo utilizando los corchetes de selección no es suficiente. Los operadores `loc`/`iloc` son requeridos enfrente de los corchetes de selección. Cuando usamos `loc`/`iloc` , la parte antes de la coma son las filas que queremos, y la parte luego de la coma son las columnas que queremos seleccionar.

Estoy interesado en las filas 10 hasta 25 y las columnas 3 hasta 5

```python
titanic.iloc[9:25, 2:5] #[9:25, 2:5] == [rango de filas, rango de columnas], el 25 y el 5 no lo incluye, es como un conjunto semiabierto
#iloc USO PARA RANGO Y NUMERO
```

De nuevo, un subconjunto de ambas, filas y columnas, no es posible en una sentada solo utilizando los corchetes de selección. Cuando nos interesa una cierta cantidad de filas basada en su posición en la tabla, utilizamos el operador `iloc` en frente a los corchetes de selección

Cuando seleccionamos filas y/o columnas específicas con `loc` o `iloc`, nuevos valores pueden ser asignados a los datos seleccionados. Por ejemplo, para asignar el nombre `anonymous` a los primeros 3 elementos de la tercera columna:

```python
titanic.iloc[0:3, 3] = "anonymous"
#[0:3, 3] primeras 3 filas y la columna numero 3 osea Name asignamos el nombre anonymus
titanic.head()
```
